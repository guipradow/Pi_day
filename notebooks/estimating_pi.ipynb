# Monte Carlo Simulation

The Monte Carlo simulation calculates a series of possible realizations of a phenomenon, along with the weight of probability of a specific occurrence, while trying to explore the whole space of the parameters of the phenomenon in question.  
Once this random sample has been calculated, the simulation gathers measurements of the quantities of interest on this sample. It is well executed if the average value of these measurements on the system realizations converge to the true value.

### Important Note

The functionality of the Monte Carlo simulation can be summarized as follows: a phenomenon is observed $n$ times, and the methods adopted in each event are recorded, with the aim of identifying the statistical distribution of the character.


## Applyng the Monte Carlo method for Pi estimation

Thee Monte Carlo method is a problem-solving strategy that uses statistics. If we indicate with $P$ the probability of a certain event, then we can randomly simulate this event and obtain $P$ by finding the ratio between the number of times our event occurred and the number of total simulations, as follows:

\begin{equation} P = \frac{\text{number of event occurrences}}{\text{number of total simulations}} \end{equation}

We can apply this strategy to get an approximation of Pi. Pi ($\pi$) is a mathematical constant indicating the relationship between the length of a circumference and its diameter. In fact, if we denote by $C$ the length of a circumference and by $d$ its diameter, we know that $C = d \times \pi$. The length of a circumference with diameter equal to 1 is worth $\pi$.

Given a circle of radius 1, it can be inscribed in a square of length 2. For convenience, we will only consider a fraction of the circle as shown in the following figure:


<p><center><figure align="middle"><img src="../figs/fraction_of_circle.png" width=25%; height=25%/></figure></center></p>

By analyzing the previous figure, we can see that the area of the square in blue is 1 and that the area of the circular sector in yellow (1/4 of the circle) is instead $\pi/4$. We randomly place a very large number of points inside the square. Thanks to the very large number an random distribution, we can approximate the size of the areas with the number of points contained in them.

If we generate $N$ random numbers inside the square, the number of points that fall in the circular sector, which we will denote by $M$, divided by the total number of generated numbers, $N$, we will have to approximate the area of the circular sector and therefor it will be equal to $\pi/4$. From this, we can derive the following equation:

\begin{equation} \pi = \frac{4M}{N} \end{equation}

The grater the number of points generated, the more precise the approximation of $\pi$ will be.

```python
N = 10000  # Numbers to be generated
M = 0  # Initializate the number of points that fall in the circular sector

X_circle = []  # List with x components of the points to be generated that fall in the circular sector
Y_circle = []  # List with y components of the points to be generated that fall in the circular sector
X_square = []  # List with x components of the points to be generated that don't fall in the circular sector
Y_square = []  # List with x components of the points to be generated that don't fall in the circular sector
```

```python
def estimate_pi(N):
    '''
    This function generate x, y random components,
    verify if the point (x, y) fall in the
    circular sector and increment M if it is
    True. The components are appended in its
    respective lists. Then, a graph with points
    is plotted.

    Parameters
    --------------------------------------------
    N: int, required
        The number of numbers to be generated
    '''
    M = 0
    for p in range(N):
        x = np.random.random()
        y = np.random.random()
        if (x**2 + y**2 <= 1):
            M += 1
            X_circle.append(x)
            Y_circle.append(y)
        else:
            X_square.append(x)
            Y_square.append(y)
    Pi = 4 * M / N
    print(f'N = {N}, M = {M}, Pi = {Pi}')

    # Draw the graph
    X_lin = np.linspace(0, 1)
    Y_lin = []
    for x in X_lin:
        Y_lin.append(np.sqrt(1 - x**2))
    plt.figure(figsize=(6.4, 6.4))
    plt.plot(X_lin, Y_lin, c='r', lw=2)
    plt.scatter(X_circle, Y_circle, color='yellow', marker='.')
    plt.scatter(X_square, Y_square, color='blue', marker='.')
    plt.title('Monte Carlo method for Pi estimation')
    plt.show()
```

```python
# Verify the estimated value of Pi for N = 10,000
N = 10_000
estimate_pi(N)
```

```python
# Verify the estimated value of Pi for N = 100,000
N = 100_000
estimate_pi(N)
```
